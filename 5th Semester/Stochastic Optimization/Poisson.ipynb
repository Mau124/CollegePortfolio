```python
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

def get_stock(alpha, lamda, theta):
    probability = 0
    s = 0

    while probability < 1 - alpha:
        probability += (math.exp(-theta*lamda)*(lamda*theta)**s)/math.factorial(s)
        s+=1

    return s-1
```

```python
print('El número de stocks necesarios es: ', get_stock(0.05, 0.01, 1000))
```

```python
probabilities = []

alpha = 0.05
theta = 0.01

for i in range(1000, 5000):
    # print(i)
    probabilities.append(get_stock(alpha, theta, i))

figure(figsize=(16, 5), dpi=80)
plt.xlabel('Tiempo')
plt.ylabel('Número de stocks')
plt.plot(probabilities)
```
