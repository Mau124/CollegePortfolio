```python
import pandas as pd
import numpy as np
import folium
```

```python
from google.colab import drive
drive.mount('/content/drive')

rutas =pd.read_csv("./drive/MyDrive/Colab Notebooks/CoordsRutas.csv")
print(f"El tamanio del data set es de {rutas.shape}")
rutas
```

```python
for label, content in rutas.items():
  coordsFinal = []
  for ruta in content:
    if isinstance(ruta, str):
      ruta = ruta.split(',')

      lat = float(ruta[0])
      lon = float(ruta[1])

      coordsFinal.append((lat, lon))

  coordsFinal = list(dict.fromkeys(coordsFinal))
  print(len(coordsFinal))
```

```python
coordsFinal = []

for ruta in rutas['Ruta 8']:
  if isinstance(ruta, str):
    ruta = ruta.split(',')
    print(ruta)

    lat = float(ruta[0])
    lon = float(ruta[1])

    coordsFinal.append((lat, lon))

coordsFinal = list(dict.fromkeys(coordsFinal))
coordsFinal
```

```python
start_latlng = (20.6489176, -103.3409776)
place=["Guadalajara, Jalisco, Mexico", "Zapopan, Jalisco, Mexico", "San Pedro Tlaquepaque, Jalisco, Mexico"]
mode="drive"
optimizer = 'time'

graph = ox.graph_from_place(place, network_type = mode)
```

```python
from folium.features import DivIcon

last_node = ox.get_nearest_node(graph, start_latlng)
node = ox.get_nearest_node(graph, coordsFinal[1])

route = nx.shortest_path(graph, last_node, node, weight=optimizer)
shortest_route_map = ox.plot_route_folium(graph, route)

for i in range(2, len(coordsFinal)):
  print(coordsFinal[i])
  node = ox.get_nearest_node(graph, coordsFinal[i])
  route = nx.shortest_path(graph, last_node, node, weight=optimizer)
  shortest_route_map = ox.plot_route_folium(graph, route, route_map=shortest_route_map)
  last_node = node

  folium.Marker(
      location=coordsFinal[i],
      icon=folium.map.Icon(color='yellow', icon="circle", prefix="fa"),
    ).add_to(shortest_route_map)

  # folium.Marker(coordsFinal[i], icon=DivIcon(
  #     icon_size=(150,36),
  #     icon_anchor=(7,20),
  #     html=f'<div style="font-size: 28pt; color : black">{i-1}</div>',
  # )).add_to(shortest_route_map)

des = ox.get_nearest_node(graph, start_latlng)
route = nx.shortest_path(graph, last_node, des , weight=optimizer)
shortest_route_map = ox.plot_route_folium(graph, route, route_map=shortest_route_map)

cedis = folium.Marker(
      location=[20.6489176, -103.3409776],
      icon=folium.Icon(color="red", icon="home", prefix="fa"),
      popup=f"<stong>CEDIS de Coppel</stong>",
      tooltip=f"CEDIS")
cedis.add_to(shortest_route_map)
```
